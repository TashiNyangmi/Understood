```python
df_lr = pd.get_dummies(df, columns = ["work", "city"], drop_first = True)
```

```python
df_lr.head()
```

Run a Logistic Regression using all columns as features

```python
model = LogisticRegression()
selected_features = df_lr.columns.tolist()
selected_features.remove("work_yes")

X = df_lr[selected_features]
y = df_lr["work_yes"]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 23)# split the data
model.fit(X_train, y_train)# train the model
y_pred = model.predict(X_test) # make prediction on test data
```

```python
# Metrics
accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test,y_pred)
recall = recall_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred)

# Print each of our scores to inspect performance.
print("Accuracy Score: %f" % accuracy)
print("Precision Score: %f" % precision)
print("Recall Score: %f" % recall)
print('F1 Score %f' % f1)
```

```python
# Some Functions
def acc (current_variables, target, basetable):
    X = basetable [current_variables]
    y = basetable [target]
    model = LogisticRegression ( )
    X_train, X_test, y_train, y_test  = train_test_split(X, y, test_size = 0.2, random_state = 23)
    model.fit (X_train, y_train)
    y_pred = model.predict(X_test)
    score = accuracy_score(y_test, y_pred)
    return (score)
```

```python
# Test above function
current_variables = ["educw", "child6", "city_yes"]
target = ["work_yes"]
score = acc(current_variables, target, df_lr) 
print (score)
```

```python
def next_best (current_variables, candidate_variables, target, basetable):
    current_score = acc(current_variables, target, basetable)
    best_variable = None
    for i in candidate_variables:
        variables = current_variables + [i]
        new_score = acc(variables, target, basetable)
        if new_score > current_score: 
            current_score = new_score
            best_variable = i
    return(best_variable)
```
